### Creating the Dataframe

```python
# Importing Pandas and NumPy and preventing a warning message
import pandas as pd
import numpy as np
pd.options.mode.chained_assignment = None
```

```python
# Scraping the table from the website
url = "https://en.wikipedia.org/wiki/List_of_London_boroughs"
dfs = pd.read_html(url)

print(len(dfs))
```

```python
# Converting the table into a Dataframe
df = dfs[0]
df
```

### Cleaning the Dataframe

```python
# Only taking the columns of interest
df = df[["Borough", "Co-ordinates"]]
df
```

```python
# Dropping the first row as there are no coordinate values
df.drop([0], inplace = True)
df.reset_index(drop = True, inplace = True)
df
```

```python
# Removing the unnecessary 'notes' in the Borough names
df.replace({"Borough": {"Greenwich [note 2]": "Greenwich", 
                        "Hammersmith and Fulham [note 4]": "Hammersmith and Fulham"}}, inplace = True)
df
```

```python
# Splitting the coordinates column to only have the needed values
df[["LatLong1", "LatLong2"]] = df["Co-ordinates"].str.split("/", expand = True)
df = df[["Borough", "LatLong2"]]
df
```

```python
# Splitting the LatLong2 column into separate latitude and longitude columns
df1 = df["LatLong2"].str.split(" ", expand = True)
df1.columns = ["LongLat{}".format(x) for x in df1.columns]
df = df.join(df1)
df = df[["Borough", "LongLat1", "LongLat2"]]

# Renaming the columns
df.rename(columns = {"LongLat1": "Latitude", 
                     "LongLat2": "Longitude"}, inplace  = True)
df
```

```python
# Removing the unnecessary '°N' from the latitude values
df["Latitude"] = df["Latitude"].str.rstrip("°N")
df
```

```python
# Adjusting the longitude values so they are in the correct format
london_df = df

london_df["Longitude"] = london_df["Longitude"].str.rstrip("°W")
london_df["Longitude"] = np.where(london_df["Longitude"].str.len() == 6, 
                                  "-" + london_df["Longitude"], 
                                  london_df["Longitude"].str.rstrip("°E"))

london_df
```

```python
# Removing hidden byte order mark from web scraping
london_df["Latitude"] = london_df["Latitude"].str.replace(u"\ufeff", "")
london_df["Longitude"] = london_df["Longitude"].str.replace(u"\ufeff", "")
```

```python
#Producing a map of London using the latitude and longitude values
map_london = folium.Map(location = [latitude, longitude], zoom_start = 10)

# Marking all the Boroughs on the map
for lat, lng, borough in zip(london_df["Latitude"], london_df["Longitude"], london_df["Borough"]):
    label = "{}".format(borough)
    label = folium.Popup(label, parse_html = True)
    folium.CircleMarker([lat, lng], 
                        radius = 5, 
                        popup = label, 
                        color = "blue", 
                        fill = True, 
                        fill_color = "3186cc", 
                        fill_opacity = 0.7, 
                        parse_html = False).add_to(map_london)

map_london
```

### Analysing the Boroughs

The countries defined as 'East Asian' (China, Hong Kong, Japan, Macau, Mongolia, North Korea, South Korea and Taiwan) would be expected to have similar cuisines to that in South Korea.

```python
# Getting all the unique venue categories to separate out the 'East Asian' restaurants
print(london_venues["Venue Category"].unique())
```

```python
# Separating out the 'East Asian' types of restaurants
east_asian_cuisine = ["Chinese Restaurant", "Asian Restaurant", "Sushi Restaurant", "Korean Restaurant", 
                      "Noodle House", "Ramen Restaurant", "Taiwanese Restaurant", "Japanese Restaurant", 
                      "Hunan Restaurant", "Okonomiyaki Restaurant", "Dumpling Restaurant", "Dim Sum Restaurant", 
                      "Cantonese Restaurant"]
```

```python
# One hot encoding each venue type for each Borough
london_onehot = pd.get_dummies(london_venues[["Venue Category"]], prefix = "", prefix_sep = "")

# Adding the Borough column to the new Dataframe
london_onehot["Borough"] = london_venues["Borough"]

# Re-ordering the columns to make sure the Borough column is first in the Dataframe
fixed_columns = [london_onehot.columns[-1]] + list(london_onehot.columns[:-1])
london_onehot = london_onehot[fixed_columns]

london_onehot
```

```python
# Grouping the rows by Borough and taking the mean of the frequency of each category
london_grouped = london_onehot.groupby(["Borough"]).mean().reset_index()
london_grouped
```

```python
# Creating a new Dataframe containing just the 'East Asian' restaurants in each Borough
london_asian_restaurants = london_grouped[["Borough", "Asian Restaurant", "Cantonese Restaurant", "Chinese Restaurant", 
                                           "Dim Sum Restaurant", "Dumpling Restaurant", "Hunan Restaurant", 
                                           "Japanese Restaurant", "Korean Restaurant", "Noodle House", "Okonomiyaki Restaurant",
                                           "Ramen Restaurant", "Sushi Restaurant", "Taiwanese Restaurant"]]
london_asian_restaurants
```

```python
# Summing up all the frequencies in each Borough
summed_restaurants = london_asian_restaurants.sum(axis = 1)
summed_restaurants
```

```python
# Adding the summed values to the main Dataframe
london_asian_restaurants["Total"] = summed_restaurants
london_asian_restaurants
```

```python
# Sorting the Dataframe so the Boroughs with the highest frequency of 'East Asian' restaurants are at the top
sorted_restaurants = london_asian_restaurants.sort_values(by = ["Total"], ascending = False, ignore_index = True)
sorted_restaurants
```

This Dataframe shows that the Boroughs with the highest proportion of 'East Asian' restaurants are Lambeth and Kingston upon Thames, both with a frequency of 0.07, followed by Croydon and Lewisham (0.0625 and 0.0513 respectively).
